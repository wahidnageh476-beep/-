```python
import pandas as pd
import numpy as np
# تحميل البيانات
df = pd.read_csv("D:\\pdf\\datesetuniversty\\Universities_Egypt.csv")
# معاينة أولية
print("شكل البيانات:", df.shape)
print("\nالأعمدة:")
print(df.columns.tolist())
df.head()
```

## 1️⃣ نظرة عامة على البيانات

```python
# معلومات عامة
df.info()
```

```python
# إحصائيات وصفية للأعمدة الرقمية
df[['Minimum Absolute Score', 'Minimum Percentage (%)']].describe().round(1)
```

```python
# القيم الفريدة في كل عمود نصي
for col in ['University_clean', 'Faculty', 'City', 'Stream', 'Stream_Type']:
    print(f"{col}: {df[col].nunique()} قيمة فريدة")
```

## 2️⃣ استعلامات أساسية

```python
# ── جميع الجامعات والمدن ──
df[['University_clean', 'City', 'Founded_Year_clean']].drop_duplicates().reset_index(drop=True)
```

```python
# ── عدد الكليات لكل جامعة ──
df.groupby('University_clean')['Faculty'].nunique().sort_values(ascending=False).reset_index()  .rename(columns={'Faculty': 'عدد الكليات'})
```

```python
# ── عدد الأقسام لكل كلية (على مستوى كل الجامعات) ──
df.groupby('Faculty')['Department'].nunique().sort_values(ascending=False).reset_index()  .rename(columns={'Department': 'عدد الأقسام'})
```

## 3️⃣ تحليل مجاميع القبول

```python
# ── أعلى وأقل مجموع لكل شعبة ──
df.groupby('Stream')['Minimum Absolute Score'].agg(['min','max','mean']).round(2)  .rename(columns={'min':'أقل مجموع','max':'أعلى مجموع','mean':'متوسط'})  .reset_index()
```

```python
# ── ترتيب الجامعات حسب المجموع — شعبة علمي رياضة ──
(
    df[df['Stream'] == 'Science-Math']
    .drop_duplicates('University_clean')
    [['University_clean','City','Minimum Absolute Score','Minimum Percentage (%)']]
    .sort_values('Minimum Absolute Score', ascending=False)
    .reset_index(drop=True)
)
```

```python
# ── ترتيب الجامعات حسب المجموع — شعبة علمي علوم ──
(
    df[df['Stream'] == 'Science-Science']
    .drop_duplicates('University_clean')
    [['University_clean','City','Minimum Absolute Score','Minimum Percentage (%)']]
    .sort_values('Minimum Absolute Score', ascending=False)
    .reset_index(drop=True)
)
```

```python
# ── ترتيب الجامعات حسب المجموع — شعبة أدبي ──
(
    df[df['Stream'] == 'Literary']
    .drop_duplicates('University_clean')
    [['University_clean','City','Minimum Absolute Score','Minimum Percentage (%)']]
    .sort_values('Minimum Absolute Score', ascending=False)
    .reset_index(drop=True)
)
```

## 4️⃣ استعلامات التصفية والفلترة

```python
# ── الكليات المتاحة في جامعة معينة (غيّر الاسم حسب ما تريد) ──
UNIVERSITY = 'cairo university'  # ← غيّر هنا

df[df['University_clean'] == UNIVERSITY]['Faculty'].unique()
```

```python
# ── الجامعات التي تقبل بمجموع أقل من قيمة معينة (علمي رياضة) ──
MIN_SCORE = 350  # ← غيّر حسب مجموعك

(
    df[(df['Stream'] == 'Science-Math') & (df['Minimum Absolute Score'] <= MIN_SCORE)]
    .drop_duplicates('University_clean')
    [['University_clean','City','Minimum Absolute Score','Minimum Percentage (%)']]
    .sort_values('Minimum Absolute Score')
    .reset_index(drop=True)
)
```

```python
# ── ابحث عن كلية معينة في كل الجامعات ──
FACULTY = 'Faculty of Medicine'  # ← غيّر هنا

(
    df[df['Faculty'] == FACULTY]
    .drop_duplicates('University_clean')
    [['University_clean','City','Stream','Minimum Absolute Score']]
    .sort_values('Minimum Absolute Score', ascending=False)
    .reset_index(drop=True)
)
```

```python
# ── جامعات مدينة معينة ──
CITY = 'Cairo'  # ← غيّر هنا

df[df['City'] == CITY][['University_clean','Faculty','Department']].drop_duplicates().reset_index(drop=True)
```

## 5️⃣ مقارنات متقدمة

```python
# ── مقارنة جامعتين في الشعب الثلاثة ──
UNI_A = 'cairo university'
UNI_B = 'ain shams university'

comparison = df[df['University_clean'].isin([UNI_A, UNI_B])]\    .groupby(['University_clean','Stream'])['Minimum Absolute Score']\    .first().unstack('Stream').round(2)

comparison
```

```python
# ── pivot table: متوسط المجموع لكل جامعة × كل شعبة ──
pivot = df.pivot_table(
    index='University_clean',
    columns='Stream',
    values='Minimum Absolute Score',
    aggfunc='mean'
).round(2)

pivot
```

```python
# ── الجامعة الأقدم والأحدث ──
age_df = df[['University_clean','Founded_Year_clean']].drop_duplicates().dropna()
print("الأقدم:", age_df.loc[age_df['Founded_Year_clean'].idxmin()].values)
print("الأحدث:", age_df.loc[age_df['Founded_Year_clean'].idxmax()].values)
```

```python
# ── عدد الأقسام لكل جامعة (الأكثر تنوعاً) ──
df.groupby('University_clean')['Department'].nunique()\  .sort_values(ascending=False)\  .reset_index()\  .rename(columns={'Department': 'عدد الأقسام'})
```

## 6️⃣ استعلام مخصص — أدخل مجموعك واعرف أقبل فين
غيّر القيم في الخلية دي حسب وضعك:

```python

#  ← غيّر القيم دي حسب حالتك       
MY_SCORE  = 380          # مجموعك
MY_STREAM = 'Science-Math'  # Science-Math | Science-Science | Literary


result = (
    df[
        (df['Stream'] == MY_STREAM) &
        (df['Minimum Absolute Score'] <= MY_SCORE)
    ]
    .drop_duplicates('University_clean')
    [['University_clean','City','Minimum Absolute Score','Minimum Percentage (%)']]
    .sort_values('Minimum Absolute Score', ascending=False)
    .reset_index(drop=True)
)

print(f"بمجموع {MY_SCORE} في شعبة {MY_STREAM}، تقدر تقبل في {len(result)} جامعة:")
result
```
